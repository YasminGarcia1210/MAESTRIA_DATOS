* ## El puerto 443 es el puerto estándar utilizado para HTTPS 
(Hypertext Transfer Protocol Secure). HTTPS es una extensión segura del protocolo HTTP (Hypertext Transfer Protocol) que se utiliza para asegurar la comunicación entre el navegador web del usuario y el servidor web.

* El puerto 8081 es comúnmente utilizado como un puerto alternativo para servicios web y aplicaciones. Aunque no tiene un uso estándar tan definido como el puerto 80 (HTTP) o el puerto 443 (HTTPS), a menudo se usa en varias situaciones:

>> Usos Comunes del Puerto 8081:
Servidor Web Alternativo:

 >> Puede usarse como un puerto alternativo para servicios HTTP o HTTPS cuando el puerto 80 o 443 está ocupado o no es accesible.
Aplicaciones de Desarrollo:

>> Se utiliza frecuentemente en entornos de desarrollo y prueba para ejecutar aplicaciones web localmente sin interferir con los puertos estándar.
Servicios de Proxy o Proxy Inverso:

>> A veces se usa para servicios de proxy, incluyendo servidores proxy inversos que redirigen el tráfico a otros servidores.
