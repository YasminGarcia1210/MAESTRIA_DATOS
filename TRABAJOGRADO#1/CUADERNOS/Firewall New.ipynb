```python
# 1. Agrupar y contar ocurrencias
df_agrupado = df.groupby(['hora', 'json.geoip_region']).size().reset_index(name='count')

# Ordenar df_agrupado por la columna 'count' en orden descendente
df_agrupado = df_agrupado.sort_values(by='count', ascending=False)

# Seleccionar los primeros 100 registros
top_100 = df_agrupado.head(80).sort_values(by='hora')

# Crear el diagrama de puntos
plt.figure(figsize=(10, 7))

# Graficar cada 'json.geoip_region' individualmente
for address in top_100['json.geoip_region'].unique():
    subset = top_100[top_100['json.geoip_region'] == address]
    plt.scatter(subset['hora'], subset['count'], label=address)

plt.xlabel('Hora')
plt.ylabel('Count')
plt.title('Diagrama de Puntos: Hora vs RegiÃ³n')
plt.legend(title='RegiÃ³n', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90, fontsize=8)  # Ajusta el tamaÃ±o de la fuente segÃºn sea necesario
plt.tight_layout()
plt.show()
```

# 2. json.destination_port

```python
# 1. Agrupar y contar ocurrencias
df_agrupado = df.groupby(['hora', 'json.destination_port']).size().reset_index(name='count')

# Ordenar df_agrupado por la columna 'count' en orden descendente
df_agrupado = df_agrupado.sort_values(by='count', ascending=False)

# Seleccionar los primeros 80 registros
top_100 = df_agrupado.head(80).sort_values(by='hora')

# Crear el diagrama de puntos
plt.figure(figsize=(10, 7))

# Graficar cada 'json.destination_port' individualmente
for port in top_100['json.destination_port'].unique():
    subset = top_100[top_100['json.destination_port'] == port]
    plt.scatter(subset['hora'], subset['count'], label=port)

plt.xlabel('Hora')
plt.ylabel('Count')
plt.title('Diagrama de Puntos: Hora vs Puerto de Destino')
plt.legend(title='Puerto de Destino', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90, fontsize=7)  # Ajusta el tamaÃ±o de la fuente segÃºn sea necesario
plt.tight_layout()
plt.show()
```

* El puerto 443 es el puerto estÃ¡ndar utilizado para HTTPS (Hypertext Transfer Protocol Secure). HTTPS es una extensiÃ³n segura del protocolo HTTP (Hypertext Transfer Protocol) que se utiliza para asegurar la comunicaciÃ³n entre el navegador web del usuario y el servidor web.

* El puerto 8081 es comÃºnmente utilizado como un puerto alternativo para servicios web y aplicaciones. Aunque no tiene un uso estÃ¡ndar tan definido como el puerto 80 (HTTP) o el puerto 443 (HTTPS), a menudo se usa en varias situaciones:

>> Usos Comunes del Puerto 8081:
Servidor Web Alternativo:

 >> Puede usarse como un puerto alternativo para servicios HTTP o HTTPS cuando el puerto 80 o 443 estÃ¡ ocupado o no es accesible.
Aplicaciones de Desarrollo:

>> Se utiliza frecuentemente en entornos de desarrollo y prueba para ejecutar aplicaciones web localmente sin interferir con los puertos estÃ¡ndar.
Servicios de Proxy o Proxy Inverso:

>> A veces se usa para servicios de proxy, incluyendo servidores proxy inversos que redirigen el trÃ¡fico a otros servidores.
