# Descripción de Variables



*  Servicio: Categorica Nominal
Esta variable obede al tipo de servicio del usuario
Los que empiezan con 6 y 7 son servicios de Internet y televisiónd e fibra optica, los que empiezan con 5 son servicio de cable

*   Suscripción: Categorica Nominal
Este variable representa el id unico del usuario que contrata el servicio, si un mismo usuario contrata diferentes servicios, el codigo de Suscripc es el mismo

* Categoria: Categorica Nominal
Esta variable represneta el tipo de srevicio que contrata el usuario, este servicio puede ser: INTERNET, TELEVISIÓN, TELEFONIA, PBX , DEDICADO

* Barrio: Categorica Nominal
Ubicación Geoespacial  de la residencia donde se instalará el o los servicios

* V.Iva: Cuantitativa continua
El impuesto IVA que va inlcuida en el precio de la factura

```python
# Agrupar por "BARRIO" y "CATEGORIA" y calcular la suma de "V.CUENTA" para cada grupo
data_4 = data.groupby(["BARRIO", "CATEGORIA"])["V.CUENTA"].sum().reset_index()

# Cambiar el nombre de la columna "V.CUENTA" a "Suma_Servicios" (opcional)
data_4.rename(columns={"V.CUENTA": "Suma_Servicios"}, inplace=True)

# Mostrar la tabla resultante
print(data_4)
```

```python
pivot_data = data_4.pivot(index="BARRIO", columns="CATEGORIA", values="Suma_Servicios")

# Graficar un gráfico de barras apiladas
pivot_data.plot(kind="line", stacked=True, figsize=(10, 6))
plt.xlabel("Barrio")
plt.ylabel("Suma de Servicios")
plt.title("Suma de Servicios por Barrio y Categoría")
plt.legend(title="Categoría")

plt.show()
```
